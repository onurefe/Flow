```python
import numpy as np
```

```python
us_freq = 40000.0
us_modulation_freq = 200.0
device_output_rate = 100.0
```

```python
low_freq = us_freq - us_modulation_freq
high_freq = us_freq + us_modulation_freq
```

```python
low_mod_signal_sin = np.sin(2 * np.pi * low_freq * np.linspace(0.0, 1.0, int(4 * us_freq / device_output_rate)))
high_mod_signal_sin = np.sin(2 * np.pi * high_freq * np.linspace(0.0, 1.0, int(4 * us_freq / device_output_rate)))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(low_mod_signal_sin)
plt.show()
```

```python
plt.plot(high_mod_signal_sin)
plt.show()
```
